# **Pareto Analysis**

```python

vendor_expenses = df.groupby('Vendor')['Amount'].sum().sort_values(ascending=False)


vendor_expenses_cumsum = vendor_expenses.cumsum() / vendor_expenses.sum() * 100

# Calculate the cumulative percentage of the number of vendors
vendor_cum_percentage = (vendor_expenses.rank(ascending=False) / len(vendor_expenses)) * 100
```

```python
import matplotlib.pyplot as plt


fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.plot(vendor_cum_percentage, vendor_expenses_cumsum, color='blue', marker='o')
ax1.set_xlabel('Cumulative Percentage of Vendors')
ax1.set_ylabel('Cumulative Percentage of Expenses', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')


ax1.axhline(80, color='gray', linestyle='--')


ax2 = ax1.twinx()
ax2.set_ylabel('Number of Vendors', color='orange')
ax2.hist(vendor_cum_percentage, bins=50, alpha=0.5, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Pareto Analysis (80/20 Rule) - Vendor Expenses')
plt.tight_layout()
plt.show()
```
